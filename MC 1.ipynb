```python
def runEpisode(ind, p1, p2, p3, p4, p5, p6, price=price, median=median, qu1=qu1, qu3=qu3):
    cash = 10000
    flag = False
    nshares = 0
    fee = 9.99
    nmax = 253
    
    for i in range(ind, (ind+nmax-1)):
        if price[i] < qu1[i] and flag == False:
            # opp to buy
            ratio = (median[i]-price[i])/(median[i]-qu1[i])
            if ratio > 3:
                p = p1
            elif ratio > 2:
                p = p2
            else: p = p3
            action = random.choices(['buy', 'wait'], [p, 1-p])
            if action[0] == 'buy':
                nshares = ((cash-fee)/price[i])//1
                cash = cash - nshares*price[i] - fee
                if nshares:
                    flag = True
                    
        elif price[i] > qu3[i] and flag == True:
            # opp to sell
            ratio = (price[i]-median[i])/(qu3[i]-median[i])
            if ratio > 3:
                p = p6
            elif ratio > 2:
                p = p5
            else: p = p4
            action = random.choices(['sell', 'wait'], [p, 1-p])
            if action[0] == 'sell':
                cash = cash + nshares*price[i] - fee
                nshares = 0
                flag = False
                
    if nshares:
        cash = cash + nshares*price[ind+nmax-1] - fee
    
    return cash - 10200
```

```python
# for a 2 year episode:

def runEpisode(ind, p1, p2, p3, p4, p5, p6, price=price, median=median, qu1=qu1, qu3=qu3):
    cash = 10000
    flag = False
    nshares = 0
    fee = 9.99
    nmax = 507
    
    for i in range(ind, (ind+nmax-1)):
        if price[i] < qu1[i] and flag == False:
            # opp to buy
            ratio = (median[i]-price[i])/(median[i]-qu1[i])
            if ratio > 3:
                p = p1
            elif ratio > 2:
                p = p2
            else: p = p3
            action = random.choices(['buy', 'wait'], [p, 1-p])
            if action[0] == 'buy':
                nshares = ((cash-fee)/price[i])//1
                cash = cash - nshares*price[i] - fee
                if nshares:
                    flag = True
                    
        elif price[i] > qu3[i] and flag == True:
            # opp to sell
            ratio = (price[i]-median[i])/(qu3[i]-median[i])
            if ratio > 3:
                p = p6
            elif ratio > 2:
                p = p5
            else: p = p4
            action = random.choices(['sell', 'wait'], [p, 1-p])
            if action[0] == 'sell':
                cash = cash + nshares*price[i] - fee
                nshares = 0
                flag = False
                
    if nshares:
        cash = cash + nshares*price[ind+nmax-1] - fee
    
    return cash - 10400
```
